```python

aprovados = 0
recuperacao = 0
reprovados = 0

qtd_alunos = int(input('Digite a quantidade de alunos: '))


for aluno in range(1, qtd_alunos + 1):

    media = float(input(f'Digite a média do aluno {aluno}: '))


    if media <= 6:
        recuperacao += 1

    elif media <= 7 and media < 10:
        aprovados += 1

    elif media <= 3:
        reprovados += 1

    else:
      print('valor inválido!')


print(f'\nAprovados: {aprovados}')
print(f'Recuperação: {recuperacao}')
print(f'Reprovados: {reprovados}')

```

```python
aprovados = 0
recuperacao = 0
reprovados = 0


qtd_alunos = int(input('Digite a quantidade de alunos: '))


for aluno in range(1, qtd_alunos + 1):

    media = float(input(f'Digite a média do aluno {aluno}: '))


    if media >= 7:
        aprovados += 1

    elif media >= 5:
        recuperacao += 1

    elif media <= 4:
        reprovados += 1


print(f'\nAprovados: {aprovados}')
print(f'Recuperação: {recuperacao}')
print(f'Reprovados: {reprovados}')
```
