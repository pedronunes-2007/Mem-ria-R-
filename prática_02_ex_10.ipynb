```python
while True:
  print('          | 1 - SOMAR | 2 - SUBTRAIR | 3 - SAIR |           ')
  print('-------------------------------------------------------------')
  menu = int(input('Escolha uma das opções acima para navegar entre o programa: '))
  print('-------------------------------------------------------------')

  if menu == 1:
    print('[CALCULADORA DE SOMA]')
    s1 = float(input('\nInsira o primeiro número da soma: '))
    s2 = float(input('Insira o segundo número da soma: '))
    soma = s1 + s2
    print(f'--> {s1} + {s2} = {soma}\n')
    print('-------------------------------------------------------------')

  elif menu == 2:
    print('[CALCULADORA DE SUBTRAÇÃO]')
    sub1 = float(input('\nInsira o primeiro número da subtração: '))
    sub2 = float(input('Insira o segundo número da subtração: '))
    subtracao = sub1 - sub2
    print(f'--> {sub1} - {sub2} = {subtracao}\n')
    print('-------------------------------------------------------------')

  elif menu == 3:
    print('\nPrograma encerrado!')
    print('=============================================================')
    break

  else:
    print('[OPÇÃO INVÁLIDA]')
    print('-------------------------------------------------------------')

```
